## Visualization of evaluation sets

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import h5py

f = h5py.File("/home/rhyang/Documents/data.h5", 'r')
```

```python
index = np.random.choice(np.arange(len(f)//2)+1, size=len(f)//2, replace=False)
print (len(index))
X_container = []
y_container = []
for j in index:
    if "training%d" % j in f:
        X = f["training%d" % j][:]
        y = f["delta%d" % j][:]
        to_use_index = np.random.choice(len(y), size=len(y), replace=False)
        X_container.append(X[to_use_index])
        y_container.append(y[to_use_index])
    else:
        continue
X_eval = np.concatenate(X_container, axis=0)

y_eval = np.concatenate(y_container, axis=0)
```

```python
plt.figure(0)
plt.scatter(X_eval[:,-1,10], y_eval)
# plt.hist(X_eval[:,-1,12], 1000)
plt.figure(1)
plt.scatter(X_eval[:,-1,11], y_eval)
# plt.hist(y_eval,1000)
# fig=plt.figure(1)
# ax=fig.add_subplot(111, projection='3d')
# ax.scatter(xs=X_eval[:, -1, 0], ys=X_eval[:, -1, 1], zs=y_eval)
# plt.plot(y_eval)
# plt.show()
```
